```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(1)

# Parameters for the first distribution
mu1 = np.array([-2, 0])
cov1 = np.array([[1, -.6], [-.6, 1]])
cov1_1 = np.array([[1,0],[0,1]])

# Parameters for the second distribution
mu2 = np.array([2, 1])
cov2 = np.array([[1, -.6], [-.6, 1]])

# Number of points to generate for each distribution
num_points = 200

# Generate points for the first distribution
points1 = np.random.multivariate_normal(mu1, cov1, num_points)
points1_1 = np.random.multivariate_normal(mu1, cov1_1, num_points)

# Generate points for the second distribution
points2 = np.random.multivariate_normal(mu2, cov2, num_points)
points2_1 = np.random.multivariate_normal(mu2, cov1_1, num_points)

# Plot the points for the first distribution
# plt.scatter(points1[:, 0], points1[:, 1], label='p(x|ω1)', alpha=0.7, c='g', marker='x', s=20)
plt.scatter(points1_1[:, 0], points1_1[:, 1], label='p(x|ω1)', alpha=0.7, c='g', marker='x', s=20)

# Plot the points for the second distribution
# plt.scatter(points2[:, 0], points2[:, 1], label='p(x|ω2)', alpha=0.7, c='m', marker='8', s=20)
plt.scatter(points2_1[:, 0], points2_1[:, 1], label='p(x|ω2)', alpha=0.7, c='m', marker='8', s=20)

# Plot the means as red points
plt.scatter(mu1[0], mu1[1], color='blue', marker='o', label='cm 1', s=100)
plt.scatter(mu2[0], mu2[1], color='red', marker='o', label='cm 2', s=100)

# Plot the decision boundary line
x_line = np.linspace(-5, 5, 100)
y_a = (1 - 8 * x_line) / 2
# y_b = (8.90625 - 7.1875*x_line)/5.3125
# y_c = (9.6 - x_line * 7.1875)/5.3125
plt.plot(x_line, y_a, color='green', label='S=eye(2) dec bound')
# plt.plot(x_line, y_b, color='orange', label='S=[1,-.6;-.6,1] dec bound')
# plt.plot(x_line, y_c, color='pink', label='l12=1,l21=0.5 dec bound')

# Set labels and title
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Bayes Classifier')

# Add legend
plt.legend()

# Set limits to zoom in
plt.xlim([-5, 5])
plt.ylim([-5, 5])

# Show the plot
plt.grid(True)
plt.show() 
```
